# 日本に上陸した台風を見える化してみた

## 台風情報の読込

```python
import pandas as pd

years = list(range(2001, 2019)) # [2001, 2002, ..., 2018]

typhoon_tbls = [pd.read_csv('data_typhoon/table{}.csv'.format(year), encoding=('cp932')) for year in years]

df_typhoon = pd.concat(typhoon_tbls)

df_typhoon = df_typhoon.reset_index() # indexを振り直して、ユニークにする。　これ大事！

display(df_typhoon.describe())
display(df_typhoon.tail())
display(df_typhoon.dtypes)
```

# 日本に上陸した台風の抽出

```python
df_landing = df_typhoon[df_typhoon['上陸'] == 1]

display(df_landing.describe())
display(df_landing.head(10))
```

```python
# 各台風が最初に上陸した時のデータ　台風進路（ライン）のツールチップで使用
first_landing = df_landing.drop_duplicates(subset='台風番号')

first_landing.head()
```

```python
landing_list = df_landing['台風番号'].unique() # 上陸した「台風番号」を納めたリスト

list_idx=[] # 各台風のindexを納めたリスト　[[台風01のindexのリスト],[台風02のindex・・・],...,[台風80のindex・・・]]
plot_typhoon=[] # プロットデータ　緯度・経度
line_typhoon=[] # 進路のデータ　経度・緯度
list_year = []

for landing_num in landing_list:
    df_lnd = df_typhoon[df_typhoon['台風番号'] == landing_num] # 上陸した台風のデータ
    
    df_idx = df_lnd.index
    df_lon = df_lnd.loc[:,'経度']
    df_lat = df_lnd.loc[:,'緯度']
    sr_year= df_lnd.iloc[0, 1]

    plot = tuple(zip(df_lat, df_lon)) # 緯度, 経度
    line = tuple(zip(df_lon, df_lat)) # 経度, 緯度
    
    list_idx.append(df_idx)
    plot_typhoon.append(plot)
    line_typhoon.append(line)
    list_year.append(sr_year)

print(len(list_idx), len(plot_typhoon), len(line_typhoon), len(list_year))
print(" 日本に上陸した台風の数：" + str(len(list_idx)) + "個　（2001〜2018年の18年間）")
```

# 地図上に日本に上陸した台風を描く



```python
import folium
import branca.colormap as cm
from IPython.display import IFrame


num_typhoon = len(list_idx) # 台風の数（2001〜2018）　18年間で、80個が上陸
color = cm.linear.Paired_11.to_step(num_typhoon).scale(0, num_typhoon-1) # 台風の数だけ描画用の色を準備

msg1 = "<h4>平成{}年　{}号<BR><center>[ {} ]</center><BR> \
        <center>{}/{:2d}/{:2d} {}時</center><BR><center>中心気圧：{}hPa</center></h4>"
msg2 = "<h4>平成{}年　{}号<BR><center>[ {} ]</center><BR><center>上陸：{}/{}/{} {}時</center></h4>"
msg3 = "<h5>H{:2d} {}号</h5>"
msg4 = '<h4><center><font color="red">★　上　陸　★</font></center><BR> \
        平成{}年　{}号<BR><center>[ {} ]</center><BR> \
        <center>{}/{:2d}/{:2d} {}時</center><BR><center>中心気圧：{}hPa</center></h4>'


'''
マップオブジェクトの生成　九州全土が表示されるように設定
'''
CENTER = [32.5, 131]
map_typhoon = folium.Map(location=CENTER, 
                         tiles='cartodbpositron', 
                         zoom_start=8, 
                         control_scale=True, 
                         prefer_canvas=True
                        )


'''
進路（ライン）上にカーソルを置くと、強調表示
'''
def highlight_function(feature):
    return {
        'fillOpacity': 0.8,
        'weight': 4,
        'fillColor': 'red',
        'color': 'red'
    }


'''
進路（ライン）の書式設定
'''
def style_function(feature):
    return {
        'fillOpacity': 0.01,
        'weight': 2,
        'fillColor': 'lightgray',
        'color': 'lightgray'
    }


'''
マーカーの半径を中心気圧によって設定
'''
def get_radius(pressure):
    p_max = 1010
    p_min =  900
    r_max = 20
    r_min =  2
    
    a = -(r_max - r_min)/(p_max - p_min)
    b = r_max - p_min * a
    
    return a * pressure + b


'''
地図上にマーカーを設置　上陸の有無で設定を変更
'''
def set_marker(idx, i, landing): 
    if landing:
        alpha = 0.8
        msg = msg4
    else:
        alpha = 0.2
        msg = msg1
    
    folium.CircleMarker(
        location=[lat, lon],
        tooltip=msg.format(
                            df_typhoon.loc[idx,'年'] - 1988, # 平成に変換
                            str(df_typhoon.loc[idx,'台風番号'])[-2:], # e.g. "1801" > "01"
                            df_typhoon.loc[idx,'台風名'],
                            df_typhoon.loc[idx,'年'], 
                            df_typhoon.loc[idx,'月'], 
                            df_typhoon.loc[idx,'日'], 
                            df_typhoon.loc[idx,'時（UTC）'], 
                            df_typhoon.loc[idx,'中心気圧'],
                           ),
        radius=get_radius(df_typhoon.loc[idx,'中心気圧']), 
        color=None,
        fill_color=color(i),
        fill_opacity=alpha
    ).add_to(feature_group)

'''
台風毎に処理　
'''
df_year = df_landing.loc[:,'年']
last_year = 0

for i, course in enumerate(plot_typhoon):
    
    if list_year[i] != last_year:
        group_name = 'Ｈ{}年　上陸数：{:2d}'.format(list_year[i] - 1988, list_year.count(list_year[i]))
        feature_group = folium.FeatureGroup(name=group_name)
    last_year = list_year[i]
    
    # マーカーでプロット
    for j, (lat, lon) in enumerate(course):
        idx = list_idx[i][j]
        # 上陸の有無で、処理を変更
        if df_typhoon.loc[idx,'上陸'] == 1:
            set_marker(idx, i, True)
        else:
            set_marker(idx, i, False)
            
    # 進路（ライン）を描く
    line = folium.GeoJson(
        data={
            'type': 'LineString',
            'coordinates': line_typhoon[i]
        },
        name=(msg3.format(
                          df_typhoon.loc[idx,'年'] - 1988,
                          str(df_typhoon.loc[idx,'台風番号'])[-2:], 
                         )), # レイヤーの名前
        overlay=True,
        style_function=style_function,
        highlight_function=highlight_function
    )
    
    # ラインのツールチップの設定　台風名＆上陸日時
    df_lnd = first_landing[first_landing['台風番号']==df_typhoon.loc[idx,'台風番号']]
    folium.Tooltip(msg2.format(
                               df_typhoon.loc[idx,'年'] - 1988,
                               str(df_typhoon.loc[idx,'台風番号'])[-2:],
                               df_typhoon.loc[idx,'台風名'],
                               df_lnd['年'].values[0],
                               df_lnd['月'].values[0],
                               df_lnd['日'].values[0],
                               df_lnd['時（UTC）'].values[0]
                              )).add_to(line)
    line.add_to(feature_group)
    feature_group.add_to(map_typhoon)

# レイヤーコントロールを設置　画面の右上にある四角いやつ
folium.LayerControl().add_to(map_typhoon)

html = 'result/5_typhoon_landing.html'
map_typhoon.save(html)
IFrame(html, width=750, height=750)
```
