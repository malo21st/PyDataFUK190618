# ヒートマップ ( HeatMap )

### 交通事故を地図にプロット

```python
import pandas as pd
 
df = pd.read_csv('data_accident/accident.csv')
df_serious = df[df.serious > 0]
data_serious = df_serious.loc[:,['latitude','longitude','serious']].values 

print("重傷事故件数：", len(data_serious))
```

```python
import folium # pip install folium
from folium.plugins import HeatMap


Fukuoka_city_hall = [33.58974488,130.401803]
heat_map = folium.Map(location=Fukuoka_city_hall)

HeatMap(data_serious).add_to(heat_map)

print("重傷事故のヒートマップ")
heat_map # Jupyter Notebook で表示させたい場合
```

```python
import pandas as pd
 
df = pd.read_csv('data_accident/accident.csv')
df_injury = df[df.injury > 0]
data_injury = df_injury.loc[:,['latitude','longitude']].values # 緯度・経度のみ！

print("軽傷事故件数：", len(data_injury))
```

```python
import folium
from folium.plugins import HeatMap
from IPython.display import IFrame


Fukuoka_city_hall = [33.58974488,130.401803]
heat_map = folium.Map(location=Fukuoka_city_hall, zoom_start=17)

HeatMap(data_injury).add_to(heat_map)

print('軽傷事故のヒートマップ\nぶっちゃけ、緯度・経度だけでも描けちゃいます！')
html = 'result/1_heat_map.html'
heat_map.save(html)
IFrame(html, width=750, height=750)
```
