# マーカークラスター ( MarkerCluster )

```python
import pandas as pd

df = pd.read_csv('data_accident/accident.csv')

df_death = df[df.serious > 0]

display(df_death.head())

locations = df_death.loc[:, ['latitude', 'longitude']].values

locations[:5]
```

```python
import folium
from folium.plugins import MarkerCluster

Fukuoka_city_hall = [33.58974488,130.401803]
_map = folium.Map(location=Fukuoka_city_hall, zoom_start=9)

MarkerCluster(
    locations=locations,
    overlay=True,
    control=True
).add_to(_map)

_map.save('result/2_marker_cluster.html')

_map
```
