# GPS（GPXフォーマット）データの見える化

```python
# https://qiita.com/oinu-poppo/items/d2977c24b04c60fd82c7
# pip install msgpack
# pip install gpxpy
import gpxpy

gpx_file = open('data_gpx/track.gpx', 'r')

gpx = gpxpy.parse(gpx_file)

data = []
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            data.append([point.latitude, point.longitude])
gpx_file.close()
```

```python
import folium

_map = folium.Map(location=[33.468, 130.515], zoom_start=13)

folium.PolyLine(
    locations = data
).add_to(_map)

_map
```
